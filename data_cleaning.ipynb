```python
new_cols = df['list-view-location'].str.split('/', expand=True)


df['city'] = new_cols[0].str.strip()
df['district'] = new_cols[1].str.strip()
df['neighbourhood'] = new_cols[2].str.strip()


print(df[['city', 'district', 'neighbourhood']].tail(50))
```

```python
# Convert Turkish characters to English and make them lowercase
def convert_to_english(text):
    if pd.isnull(text):
        return text
    char_map = {
        'ç': 'c', 'ğ': 'g', 'ı': 'i', 'İ': 'i', 'ö': 'o', 'ş': 's', 'ü': 'u',
        'Ç': 'c', 'Ğ': 'g', 'Ö': 'o', 'Ş': 's', 'Ü': 'u'
    }
    for tr_char, en_char in char_map.items():
        text = text.replace(tr_char, en_char)
    return text.lower() # Lower function

# Apply columns to this function
target_cols = ['city', 'district', 'neighbourhood']

for col in target_cols:
    # Apply the conversion function to each column
    df[col] = df[col].apply(convert_to_english)

```

```python
df['neighbourhood'] = df['neighbourhood'].str.replace(r'\s+mah.*', '', regex=True).str.strip()


print(df[['city', 'district', 'neighbourhood']].head())
```
